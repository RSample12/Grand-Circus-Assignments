## Part 5: Outlier Detection and Removal


```python
# Question 1: Print out the Outliers in each column in the pima_fixed_columns_df dataframe, use the IQR method of outlier detection.
new_df = pima_missing_fixed_df
for col, series in pima_missing_fixed_df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].items():
    print(col)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    outliers = (series > (Q1 - 1.5 * IQR)) & (series < (Q3 + 1.5 * IQR))
    new_df[col] = new_df[col][outliers]
    print(outliers)
```

```python
# Question 2: Use loc to remove outliers in each of the columns that have outliers, save this as pima_outlier_removed_df.
pima_outlier_removed = new_df
pima_outlier_removed
```
