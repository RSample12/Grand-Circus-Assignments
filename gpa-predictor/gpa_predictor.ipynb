```python
import pandas as pd

# import data
df = pd.read_csv('sat_gpa_v1.csv')
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# scatterplot
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='sat_sum', y='fy_gpa', alpha=.5)
plt.title('SAT vs GPA')
plt.xlabel('SAT SCORE')
plt.ylabel('GPA')
plt.show()
```

```python
# using statsmodels create linear reg
import statsmodels.api as sm

# get X and Y
X = df['sat_sum']
y = df['fy_gpa']

# Add a constant
X = sm.add_constant(X)
X
```

```python
# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Get the model's summary
print(model.summary())
```

```python
sns.lmplot(data=df, x='sat_sum', y='fy_gpa', height=6, aspect=2, scatter_kws={'alpha':0.3})
plt.title('SAT vs GPA')
plt.xlabel('SAT SCORE')
plt.ylabel('GPA')
plt.show()
```
