```python
# Question 2:  Print the highest and lowest priced homes in Ames. \
df = df.sort_values(by='SalePrice')
df.tail(1)
```

```python
df.head(1)
```

```python
# Question 3:  How many categories does lot size have?
df['Lot Shape'].nunique()
```

```python
# Question 4:  What is the range of years for the house listings in the data?
max_yr = df['Year Built'].max()
min_yr = df['Year Built'].min()
print(f'{min_yr}-{max_yr}')
```

```python
# Question 5:  What is the total number of all homes sold in Ames for all years?
df['Yr Sold'].count()
```

```python
# Question 6:  How many nulls or NA are there in the Lot Frontage column?
df['Lot Frontage'].isna().sum()
```

```python
# Question 7:  How many items have a value in Alley and what is the total number of  NA in the column?
num_na = df['Alley'].isna().sum()
print('num values:',len(df['Alley']) - num_na)
print('number NA:', num_na)
```

```python
# Question 8:  What are all the homes in Gilbert with Yr Sold being 2009 and 2010?
new_df = df[(df['Yr Sold'] == 2009) | (df['Yr Sold'] == 2010)]
new_df
```

```python
# Question 9:  What is the count of homes sold in Gilbert with Yr Sold being 2010?
new_df['Yr Sold'].where(new_df['Yr Sold'] == 2010).count()
```

```python
# Question 10:  Return the value of the third row, and fifth column.
print(df.iloc[2]) # 3rd row
print()
print(df.iloc[:, 4]) # 5th col
```

```python
# Question 11:  Remake a new column from the SalePrice that includes data formatted with the $, and commas for the thousands/millions digits.
df["Formatted_Price"] = df["SalePrice"].map('${:,d}'.format)
df
```

```python
# Question 12:   Output a dataframe with only SalePrice, Neighborhood, and Year Built.  
# Save this file in your git repository as “minimal_ames_housing.csv”,  
# Make sure you include the formatted SalePrice from the previous question.
final_df = df[['Formatted_Price', 'Neighborhood', 'Year Built']]
final_df.to_csv('minimal_ames_housing.csv', index=False)
final_df
```
